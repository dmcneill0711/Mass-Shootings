```python
# Import Dependencies
import pandas as pd
```

```python
# Upload CSV Downloaded From GVA.org
children_killed = "Resources/Children_Killed.csv"
children_df = pd.read_csv(children_killed)
children_df
```

```python
# Drop Operations Column With All Null Values
children_df = children_df.drop(['Operations'], axis=1)

# Format The Date To mm/dd/yyyy
children_df['Incident Date'] = children_df['Incident Date'].str.strip().str.replace(' ','')
children_df['Incident Date'] = children_df['Incident Date'].str.strip().str.replace('January','01/')
children_df['Incident Date'] = children_df['Incident Date'].str.strip().str.replace('February','02/')
children_df['Incident Date'] = children_df['Incident Date'].str.strip().str.replace('March','03/')
children_df['Incident Date'] = children_df['Incident Date'].str.strip().str.replace('April','04/')
children_df['Incident Date'] = children_df['Incident Date'].str.strip().str.replace('May','05/')
children_df['Incident Date'] = children_df['Incident Date'].str.strip().str.replace('June','06/')
children_df['Incident Date'] = children_df['Incident Date'].str.strip().str.replace('July','07/')
children_df['Incident Date'] = children_df['Incident Date'].str.strip().str.replace('August','08/')
children_df['Incident Date'] = children_df['Incident Date'].str.strip().str.replace('September','09/')
children_df['Incident Date'] = children_df['Incident Date'].str.strip().str.replace('October','10/')
children_df['Incident Date'] = children_df['Incident Date'].str.strip().str.replace('November','11/')
children_df['Incident Date'] = children_df['Incident Date'].str.strip().str.replace('December','12/')
children_df['Incident Date'] = children_df['Incident Date'].str.strip().str.replace(',','/')

# Target Incident Date Column And Select Rows That Have 2022
Children_2022 = children_df.loc[children_df['Incident Date'].str.contains('2022')]

# Final DataFrame
Children_2022
```
